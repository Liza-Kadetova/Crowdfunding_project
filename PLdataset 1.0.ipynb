```
import pandas as pd
import pdb
```

```
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException

import time
from bs4 import BeautifulSoup
import re
```

```
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',options=options)
driver.maximize_window()
```

```
links = []
for i in range(1,166):
  url='https://planeta.ru/search/projects?category=CHARITY&status=SUCCESSFUL&page=' + str(i)
  links.append(url)
print(links)
len(links)
```

```

#файл записан 

myfile = open ("planeta-3.txt", 'a')
myfile
for page in links:
  driver.get(page)
  time.sleep(3)
  content = driver.page_source
  myfile.write(content)
  time.sleep(3)
myfile.close()
myfile = open("planeta-3.txt", 'r')
myfile.read()
myfile.close()
```

```
projects=[]
with open ("planeta-3.txt", 'r') as myfile2:
  myfile2 = myfile2.read()
  result=re.findall(r'\/campaigns\/.[^\"]+', myfile2)
  for i in result:
    projects.append("https://planeta.ru"+i)
  print(projects)
  print(len(projects))
```

```
def get_title(url):
  driver.get(url)
  wait = WebDriverWait(driver, 15)
  try:
    x=wait.until(EC.presence_of_element_located((By.XPATH, '//h1[@class="project-view_name ng-star-inserted"]'))).text
  except:
    driver.refresh()
    try:
      x=wait.until(EC.presence_of_element_located((By.XPATH, '//h1[@class="project-view_name ng-star-inserted"]'))).text
    except:
      x = "NaN"
  return(x)



def get_briefly(url):
  driver.get(url)
  wait = WebDriverWait(driver, 15)
  try:
    x=wait.until(EC.presence_of_element_located((By.XPATH, '//p[@class="project-view_short-desc ng-star-inserted"]'))).text
  except:
    driver.refresh()
    try:
      x=wait.until(EC.presence_of_element_located((By.XPATH, '//p[@class="project-view_short-desc ng-star-inserted"]'))).text
    except:
      x = "NaN"
  return(x)



def get_description(url):
  driver.get(url)
  wait = WebDriverWait(driver, 15)
  try:
    x=wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="project-description sub-content ng-star-inserted"]'))).text
  except:
    driver.refresh()
    try:
      x=wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="project-description sub-content ng-star-inserted"]'))).text
    except:
      x = "NaN"
  return(x)


def get_result(url):
  driver.get(url)
  wait = WebDriverWait(driver, 15)
  try:
    x=wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="project-view-meta_amount-val ng-star-inserted"]'))).text
  except:
    driver.refresh()
    try:
      x=wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="project-view-meta_amount-val ng-star-inserted"]'))).text
    except:
      x = "NaN"
  return(x)


def get_goal(url):
  driver.get(url)
  wait = WebDriverWait(driver, 15)
  try:
    x=wait.until(EC.presence_of_element_located((By.XPATH, '//span[@class="project-view-meta_amount-lbl-val"]'))).text
  except:
    driver.refresh()
    try:
      x=wait.until(EC.presence_of_element_located((By.XPATH, '//span[@class="project-view-meta_amount-lbl-val"]'))).text
    except:
      x = "NaN"
  return(x)


def get_donations(url):
  driver.get(url)
  wait = WebDriverWait(driver, 15)
  try:
    x=wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="project-view-info_lbl"][text() ="Поддержали"]/preceding-sibling::*'))).text
  except:
    driver.refresh()
    try:
      x=wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="project-view-info_lbl"][text() ="Поддержали"]/preceding-sibling::*'))).text
    except:
      x = "NaN"
  return(x)


def get_start(url):
  driver.get(url)
  wait = WebDriverWait(driver, 15)
  try:
    x=wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="project-view-info_lbl"][text() ="Запущен"]/preceding-sibling::*'))).text
  except:
    driver.refresh()
    try:
      x=wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="project-view-info_lbl"][text() ="Запущен"]/preceding-sibling::*'))).text
    except:
      x = "NaN"
  return(x)


def get_finish(url):
  driver.get(url)
  wait = WebDriverWait(driver, 15)
  try:
    x=wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="project-view-info_lbl"][text() ="Завершен"]/preceding-sibling::*'))).text
  except:
    driver.refresh()
    try:
      x=wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="project-view-info_lbl"][text() ="Завершен"]/preceding-sibling::*'))).text
    except:
      x = "NaN"
  return(x)
```

```
DataPl = {
  "Title": [],
  "Briefly": [],
  "Description": [],
  "Result": [],
  "Goal": [],
  "Donations": [],
  "Start": [],
  "Finish": []
  }
```

```
for i in projects:
  title = get_title(i)
  briefly = get_briefly(i)
  description = get_description(i)
  result = get_result(i)
  goal = get_goal(i)
  donations = get_donations(i)
  start = get_start(i)
  finish = get_finish(i)
  DataPl["Title"].append(title)
  DataPl["Briefly"].append(briefly)
  DataPl["Description"].append(description)
  DataPl["Result"].append(result)
  DataPl["Goal"].append(goal)
  DataPl["Donations"].append(donations)
  DataPl["Start"].append(start)
  DataPl["Finish"].append(finish)
  print(title)

#print(DataPl)
```

```
PLdataset = pd.DataFrame(DataPl)
PLdataset
```

```
PLdataset.to_csv('PLdataset.csv')
```
